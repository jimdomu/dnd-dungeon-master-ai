Testing TTS Open Ai

```python
from openai import OpenAI

client = OpenAI()

response = client.audio.speech.create(
    model="tts-1",
    voice="alloy",
    input="Hello world! This is a streaming test.",
)

response.stream_to_file("output.mp3")
```

```python
import IPython
IPython.display.Audio("output.mp3", autoplay=True) #NOTE: AUTOPLAY doesn't seem to work without user interaction, so this is the furthest we can go
```
